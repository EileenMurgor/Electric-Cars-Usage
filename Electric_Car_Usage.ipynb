Data Cleaning

Validity

```python
# Procedure 1: Irrelevant Data Observation
# Data Cleaning Action: Dropping geo point , displayed comment , and scheduled at attributes
# Explanation:No question to be answered requires this columns.
#
df.drop(["Geo point" , "Displayed comment" ,"Scheduled at"], axis = 1, inplace = True)
df.head(10)
```

```python
#Procedure 2 : Syntax Errors -  Checking for syntax errors of leading or tailing white spaces in the dataset

df['Address'] = df['Address'].str.strip()
df['Address']
df['ID'] = df['ID'].str.strip()
df['ID']
```

```python
#Removing ©- the copyright sign from words in the address column
df['Address'] = df['Address'].str.replace("©","")
df['Address']
```

```python
#Renaming column names according to the pascal case
df.columns=['Address' , 'Cars' , 'BlueCarCounter' ,'UtilibCounter' , 'Utilib1.4Counter' , 'ChargeSlots' , 'ChargingStatus' , 'City' , 'Id' , 'Kind' , 'PostalCode' , 'PublicName' , 'RentalStatus' , 'Slots' , 'StationType' , 'Status' , 'SubscriptionStatus' , 'Year' , 'Month' , 'Day' , 'Hour' , 'Minute']
df.columns.values.tolist()
```

```python
#Data type conversion
#Converting the month,day,year,time columns to date time and combining them
df['Date'] =pd.to_datetime(df[['Year', 'Month', 'Day', 'Hour', 'Minute']])
df
#Dropping the minute,hour,day,month, and year columns now that we have a cobmined column
df.drop(["Year" , "Month" ,"Day" , "Hour" , "Minute"], axis = 1, inplace = True)
df.head(10)
```

```python
#Most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.
#To compute usage, we will need to understand that we have to join successive (in time) measures/counters 
#for a given station, as the difference will tell whether a car was picked up, returned, or nothing happened.
#x=df['Date'].groupby(df['PublicName'])
#x.sum()
#most_populated=df.groupby(['PublicName']).sum().sort_values('Total' , ascending=0).head(5)
# 
y=df[(df['City']== 'Paris')]
grouped = y.groupby('PublicName')
grouped = y.groupby('Date', axis='columns')
grouped = y.groupby(['PublicName', 'Date' , 'BlueCarCounter'])
grouped['BlueCarCounter'].sum()
```
